## Import Packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import Data:

```python
health = pd.read_csv("healthstats.csv")
health.info()
```

```python
fitness = pd.read_csv("workoutstats.csv")
fitness.info()
```

## Create "date" column for merging later

Adjust Health "date" column to be in date_time format.

```python
health["Date"] = pd.to_datetime(health["Date"])
health.head()
```

```python
fitness["Start"] = pd.to_datetime(fitness["Start"])
fitness.head()
```

Extract the date portion of the "start" column and create a new feature called date

```python
fitness["Date"] = fitness["Start"].dt.date
fitness.head()
```

Remove Start and End columns from fitness data. 
- These colummns have no bearing on the data. 

```python
fitness.drop(["Start","End"],axis = 1, inplace=True)
fitness.head()
```

## Merge Data

First change fitness column "date" to datetime since it is currently an object. 

```python
fitness["Date"] = pd.to_datetime(fitness["Date"])
```

```python
df = pd.merge(health, fitness, on = "Date")
df.head()
```

# Export merged DF as CSV

```python
df.to_csv('fitdata.csv', index = False) # remove index column from the csv 
```
